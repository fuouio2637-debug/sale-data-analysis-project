```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('sales_data.csv')
```

```python
print(df.isnull().sum())
```

### How many sales will show

```python
total_sales=df['Unit_Cost'].sum()
print(f'total_sales: {total_sales}')
```

```python
top_profit_products=df.groupby('Product')['Profit'].sum().sort_values(ascending=False)
print('top_5_products:')
print(top_profit_products.head(5))
```

##### what products have most sales in any categories

```python
category_revenue=df.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)
print('Revenue by Category')
category_revenue
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(data=df,x='Unit_Price',y='Profit',hue='Product_Category')
plt.title('Profit vs Unit_Price')
plt.show()
```

```python
df['Product_Margin%']=df['Profit']/df['Revenue']*100
print(df[['Product','Product_Margin%']].sort_values(by='Product_Margin%',ascending=False).head())
```

### top 3 revenue

```python
top_3_revenue=df.nlargest(3,'Revenue')
high_profit=df[df['Profit']>2000]
print(top_3_revenue)
```

#### which sub_category have most sales

```python
best_selling_sub=df.groupby('Sub_Category')['Order_Quantity'].sum().sort_values(ascending=False)
print(best_selling_sub)
```

```python
plt.figure(figsize=(14,6))
sns.barplot(data=df,x='Product_Category',y='Profit',estimator=sum)
plt.title('Profit by category')
plt.show()
```
